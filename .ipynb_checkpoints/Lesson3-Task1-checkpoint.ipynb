```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
k = 2
a = math.pi/6
b = 1
x = np.linspace(-10*np.pi, 10*np.pi, 201)
plt.plot(x, (k * np.cos(x - a) + b))
k = 2
a = math.pi/2
b = 3
plt.plot(x, (k * np.cos(x - a) + b))
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
