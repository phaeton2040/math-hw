```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X
X, Y = np.meshgrid(X, Y)
Z = 2*X + 2*Y
Z2 = 2*X + 2*Y - 15
ax.plot_surface(X, Y, Z, edgecolor='red', alpha=0.3)
ax.plot_surface(X, Y, Z2, edgecolor='green', alpha=0.3)
ax.scatter(0, 0, 0, 'z', 50, 'purple')
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
Z1 = np.sqrt(X**2 + Y**2)
Z2 = X**2 - Y**2
Z3 = X**3 - Y**2
ax.plot_surface(X, Y, Z1, edgecolor='red', alpha=0.1)
ax.plot_surface(X, Y, Z2, edgecolor='green', alpha=0.1)
ax.plot_surface(X, Y, Z3, edgecolor='blue', alpha=0.1)
ax.scatter(0, 0, 0, 'z', 50, 'purple')
```
