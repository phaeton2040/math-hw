```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

### Нарисуем круг

```python
x1  = []
x2  = []
y1  = []
y2  = []
R   = 1
for i in range(10000):
    x_ = i / 10000
    x1.append(x_)
    x2.append(-1 * x_)
    y1.append(math.sqrt(R*R - x_*x_))
    y2.append(-1 * math.sqrt(R*R - x_*x_))
plt.axes().set_aspect('equal')
plt.plot(x1, y1)
plt.plot(x1, y2)
plt.plot(x2, y1)
plt.plot(x2, y2)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Нарисуем эллипс

```python
x1  = []
x2  = []
y1  = []
y2  = []
a   = 1
b   = 2
for i in range(10000):
    x_ = i / 10000
    x1.append(x_)
    x2.append(-1 * x_)
    y1.append(math.sqrt(b*b * (1 - x_*x_/a*a)))
    y2.append(-1 * math.sqrt(b*b * (1 - x_*x_/a*a)))
plt.axes().set_aspect('equal')
plt.plot(x1, y1)
plt.plot(x1, y2)
plt.plot(x2, y1)
plt.plot(x2, y2)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Нарисуем гиперболу

```python
x1  = []
x2  = []
y1  = []
y2  = []
a   = 1
b   = 1
for i in range(10000):
    x_ = i / 10000
    x1.append(x_)
    x2.append(-1 * x_)
    y1.append(math.sqrt(b*b * (1 + x_*x_/a*a)))
    y2.append(-1 * math.sqrt(b*b * (1 + x_*x_/a*a)))
plt.axes().set_aspect('equal')
plt.plot(x1, y1)
plt.plot(x1, y2)
plt.plot(x2, y1)
plt.plot(x2, y2)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
