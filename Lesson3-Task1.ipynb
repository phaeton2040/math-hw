```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 2
a = 3*np.pi/2
b = 0.3
x = np.linspace(-2*np.pi, 2*np.pi, 201)
plt.plot(x, (k * np.cos(x - a) + b))
k = 4
a = np.pi/4
b = 0.6
plt.plot(x, (k * np.cos(x - a) + b))
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
